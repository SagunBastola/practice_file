```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
census=pd.read_csv("/content/census-income .csv")
```

```python
census.info()
```

```python
g=sns.catplot(kind="count",x="education",data=census,hue="sex")
g.fig.suptitle("Education count interms of sex",y=1.03)
g.set(xlabel="Education Type",ylabel="No of People")
plt.xticks(rotation=90)
plt.show()
```
